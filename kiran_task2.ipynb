```python
import os 

# importing io from skimage 
import skimage as sk
from skimage import io 
import cv2

# way to load image from file 
file = os.path.join(r"C:\Users\CSC\Desktop\cats.jpg") 
  
  
img = cv2.imread(file) 
io.imshow(img)
io.show()
# way to show the input image 
cv2.imshow('Original image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()  
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
io.imshow(gray)
io.show()
cv2.imshow('Grayscale',gray)
cv2.waitKey(0)
cv2.destroyAllWindows() 
cv2.imwrite(r"C:\Users\CSC\Desktop\task2\newcat.jpg",gray)
gray[:,150]=0
gray[:,50]=0
gray[100,:]=0
gray[160,:]=0

io.imshow(gray)
io.show()
cv2.imshow('Grayscale2',gray)
cv2.waitKey(0)
cv2.destroyAllWindows() 
cv2.imwrite(r"C:\Users\CSC\Desktop\task2\newcatwithlines.jpg",gray)
```

```python
import cv2
import numpy as np

from skimage import io

image = cv2.imread(r"C:\Users\CSC\Desktop\cats.jpg")

blue = (255, 0, 0)
red = (0, 0, 255 )
green = (0, 255, 0)
white = (255, 255, 255)
cv2.line(image, (0, 30), (288, 30), red, thickness=2)
cv2.line(image, (0, 80), (288, 80), green, thickness=2)
cv2.line(image, (0, 120), (288, 120), blue, thickness=2)
# To drawvertical lines
cv2.line(image, (30, 0), (30, 175), red, thickness=2)
cv2.line(image, (80, 0), (80, 175), green, thickness=2)
cv2.line(image, (120, 0), (120, 175), blue, thickness=2)

io.imshow(image)
io.show()
cv2.imshow("cat", image)

cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite(r"C:\Users\CSC\Desktop\check\newcatcolor1.jpg",image)
```

```python
import numpy as np
import cv2
from skimage import io

img = cv2.imread(r"C:\Users\CSC\Desktop\cats.jpg") 

# way to show the input image 
cv2.imshow('Origianl image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()


# creating a rectangle
r=np.shape(img)
cv2.rectangle(img, (80, 30), (135, 140), (0, 0, 255), 2)
cv2.rectangle(img, (150, 50), (255, 140), (255, 0, 0), 2)
io.imshow(img)
io.show()
cv2.imshow('Bounding box',img)
# allows us to see image
# until closed forcefully
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite(r"C:\Users\CSC\Desktop\check\boundingbox.jpg",img)
```
